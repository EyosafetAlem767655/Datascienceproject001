## This is a web scrapping model for tesla and Gamestop shop

```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
# 1. Use yfinance to Extract Stock Data
tesla_ticker = yf.Ticker("TSLA.MX") 
tesla_data = tesla_ticker.history(period="1y")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'

tables = pd.read_html(url)

if len(tables) > 0:
    tesla_revenue = tables[0]
    print(tesla_revenue.tail())
else:
    print("No tables found on the webpage.")
```

```python
data  = requests.get(url).text
soup = BeautifulSoup(data, 'html5lib')
```

```python
# 3. Use yfinance to Extract Stock Data (GME - GameStop)
gme_ticker = yf.Ticker("GME")
gme_data = gme_ticker.history(period="1y")
gme_data.reset_index(inplace=True)
gme_data.head()

```

```python
url2 = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html'

tables = pd.read_html(url2)

if len(tables) > 0:
    gme_revenue = tables[0]
    print(gme_revenue.tail())
else:
    print("No tables found on the webpage.")
```

```python
# 5. Plot Tesla Stock Graph
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'])
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

make_graph(tesla_data, 'Tesla Stock Data')
```

```python
# 6. Plot GameStop Stock Graph
make_graph(gme_data, 'GameStop Stock Data')
```
